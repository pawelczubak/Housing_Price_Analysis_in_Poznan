```python
import pandas as pd

df = pd.read_csv('../data/processed/output_data.csv')

print(df.head())
print(df.info())
```

```python
import matplotlib.pyplot as plt

# Grupowanie danych i liczenie liczby mieszkań w każdej dzielnicy
district_counts = df['District'].value_counts()

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
bars = plt.bar(district_counts.index, district_counts.values, color='skyblue', edgecolor='blue')

# Dodanie etykiet danych nad każdym słupkiem
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Pozycja X etykiety
        height,                             # Pozycja Y etykiety
        str(height),                        # Tekst etykiety
        ha='center',                        # Wyrównanie poziome
        va='bottom',                        # Wyrównanie pionowe
        fontsize=7                         # Rozmiar czcionki
    )

# Dostosowanie wykresu
plt.title('Liczba dostępnych mieszkań w każdej dzielnicy')
plt.xlabel('Dzielnica')
plt.ylabel('Liczba mieszkań')
plt.xticks(rotation=90)
plt.tight_layout()

# Wyświetlenie wykresu
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Przykładowy DataFrame (wczytaj swój plik CSV)
# df = pd.read_csv('your_file.csv')

# Konwersja kolumny 'Price' na typ numeryczny
df['Price'] = df['Price'].replace('[^\d.]', '', regex=True).astype(float)

# Grupowanie danych po 'District', obliczanie średnich cen i sortowanie malejąco
average_prices = df.groupby('District')['Price'].mean().sort_values(ascending=False)

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
bars = plt.bar(average_prices.index, average_prices.values, color='skyblue', edgecolor='blue')


# Dostosowanie wykresu
plt.title('Średnie ceny mieszkań w każdej dzielnicy', fontsize=14, fontweight='bold')
plt.xlabel('Dzielnica', fontsize=12)
plt.ylabel('Średnia cena mieszkań (zł)', fontsize=12)
plt.xticks(rotation=90, fontsize=10)  # Obrót etykiet na osi X o 90 stopni
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Wyświetlenie wykresu
plt.show()
```

```python
# Konwersja kolumny 'Price_per_m' na typ numeryczny
df['Price_per_m'] = df['Price_per_m'].replace('[^\d.]', '', regex=True).astype(float)

# Grupowanie danych po 'District', obliczanie średnich cen za metr kwadratowy i sortowanie malejąco
average_price_per_m = df.groupby('District')['Price_per_m'].mean().sort_values(ascending=False)

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
bars = plt.bar(average_price_per_m.index, average_price_per_m.values, color='skyblue', edgecolor='blue')



# Dostosowanie wykresu
plt.title('Średnia cena za metr kwadratowy w każdej dzielnicy', fontsize=14, fontweight='bold')
plt.xlabel('Dzielnica', fontsize=12)
plt.ylabel('Średnia cena za metr kwadratowy (zł)', fontsize=12)
plt.xticks(rotation=90, fontsize=10)  # Obrót etykiet na osi X o 90 stopni
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Wyświetlenie wykresu
plt.show()
```

```python
# Konwersja kolumn 'Price' i 'Area' na typ numeryczny
df['Price'] = df['Price'].replace('[^\d.]', '', regex=True).astype(float)
df['Area'] = df['Area'].astype(float)

# Usunięcie wartości odstających
filtered_df = df[(df['Area'] <= 200) & (df['Price'] <= 3_000_000)]

# Tworzenie wykresu punktowego
plt.figure(figsize=(12, 6))
plt.scatter(filtered_df['Area'], filtered_df['Price'], color='blue', alpha=0.6, edgecolors='k')

# Dodanie tytułu i etykiet osi
plt.title('Zależność ceny mieszkania od powierzchni (bez wartości odstających)', fontsize=14, fontweight='bold')
plt.xlabel('Powierzchnia mieszkania (m²)', fontsize=12)
plt.ylabel('Cena mieszkania (zł)', fontsize=12)

# Dostosowanie osi
plt.grid(alpha=0.3)
plt.tight_layout()

# Wyświetlenie wykresu
plt.show()
```

```python
# Usunięcie wartości odstających (ceny powyżej 3 000 000 PLN)
df_filtered = df[df["Price"] <= 3_000_000]

# Tworzenie histogramu cen mieszkań
plt.figure(figsize=(12, 6))
plt.hist(df_filtered["Price"], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel("Cena mieszkań (PLN)")
plt.ylabel("Liczba ofert")
plt.title("Rozkład cen mieszkań w Poznaniu (bez wartości odstających)")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ustawienie osi X w pełnym formacie
plt.xticks(rotation=0)  # Obrót etykiet na osi X dla lepszej czytelności
plt.ticklabel_format(style='plain', axis='x')  # Wyłączenie notacji naukowej

plt.show()
```

```python
# Usunięcie wartości odstających (ceny powyżej 3 000 000 PLN)
df_filtered = df[df["Price"] <= 3_000_000]

# Grupowanie danych: obliczenie średniej ceny dla każdego piętra
average_price_per_floor = df_filtered.groupby("Floor")["Price"].mean().reset_index()

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
bars = plt.bar(average_price_per_floor["Floor"], average_price_per_floor["Price"], color='skyblue', edgecolor='black')


# Etykiety i tytuł wykresu
plt.xlabel("Piętro")
plt.ylabel("Średnia cena mieszkań (PLN)")
plt.title("Średnia cena mieszkań w zależności od piętra")
plt.xticks(average_price_per_floor["Floor"], rotation=0)  # Wyświetlanie wszystkich pięter na osi X
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='y')  # Wyłączenie notacji naukowej dla cen

plt.show()
```

```python
# Usunięcie wartości odstających (ceny powyżej 3 000 000 PLN)
df_filtered = df[df["Price"] <= 3_000_000]

# Wybór mieszkań, które znajdują się na ostatnim piętrze budynku
last_floor_apartments = df_filtered[df_filtered["Floor"] == df_filtered["Floors number"]]

# Grupowanie danych: obliczenie średniej ceny dla każdego ostatniego piętra
average_price_last_floors = last_floor_apartments.groupby("Floors number")["Price"].mean().reset_index()

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
bars = plt.bar(average_price_last_floors["Floors number"], average_price_last_floors["Price"], color='skyblue', edgecolor='black')


# Etykiety i tytuł wykresu
plt.xlabel("Liczba pięter w budynku (ostatnie piętro)")
plt.ylabel("Średnia cena mieszkań (PLN)")
plt.title("Średnia cena mieszkań na ostatnich piętrach")
plt.xticks(average_price_last_floors["Floors number"], rotation=0)  # Wyświetlanie wszystkich pięter na osi X
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='y')  # Wyłączenie notacji naukowej dla cen

# Wyświetlenie wykresu
plt.show()
```

```python
# Usunięcie wartości odstających (ceny powyżej 3 000 000 PLN)
df_filtered = df[df["Price"] <= 3_000_000]

# Wybór mieszkań jednopokojowych
one_room_apartments = df_filtered[df_filtered["Room number"] == 1]

# Obliczenie średniej ceny mieszkań jednopokojowych w każdej dzielnicy
average_price_one_room_per_district = one_room_apartments.groupby("District")["Price"].mean().reset_index()

# Sortowanie wyników dla lepszej czytelności
average_price_one_room_per_district = average_price_one_room_per_district.sort_values(by="Price", ascending=False)

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
bars = plt.barh(average_price_one_room_per_district["District"], average_price_one_room_per_district["Price"], color='skyblue', edgecolor='black')

# Dodanie etykiet na słupkach
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f"{int(bar.get_width()):,}".replace(",", " "), ha='left', va='center', fontsize=10)

# Etykiety i tytuł wykresu
plt.xlabel("Średnia cena mieszkań jednopokojowych (PLN)")
plt.ylabel("Dzielnica")
plt.title("Średnia cena mieszkań jednopokojowych w każdej dzielnicy Poznania")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='x')  # Wyłączenie notacji naukowej dla cen

# Wyświetlenie wykresu
plt.show()
```

```python
# Usunięcie wartości odstających (ceny powyżej 3 000 000 PLN)
df_filtered = df[df["Price"] <= 3_000_000]

# Wybór mieszkań dwupokojowych
two_room_apartments = df_filtered[df_filtered["Room number"] == 2]

# Obliczenie średniej ceny mieszkań dwupokojowych w każdej dzielnicy
average_price_two_room_per_district = two_room_apartments.groupby("District")["Price"].mean().reset_index()

# Sortowanie wyników dla lepszej czytelności
average_price_two_room_per_district = average_price_two_room_per_district.sort_values(by="Price", ascending=False)

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
bars = plt.barh(average_price_two_room_per_district["District"], average_price_two_room_per_district["Price"], color='skyblue', edgecolor='black')

# Dodanie etykiet na słupkach
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f"{int(bar.get_width()):,}".replace(",", " "), ha='left', va='center', fontsize=8)

# Etykiety i tytuł wykresu
plt.xlabel("Średnia cena mieszkań dwupokojowych (PLN)")
plt.ylabel("Dzielnica")
plt.title("Średnia cena mieszkań dwupokojowych w każdej dzielnicy Poznania")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='x')  # Wyłączenie notacji naukowej dla cen

# Wyświetlenie wykresu
plt.show()
```

```python
# Usunięcie wartości odstających (ceny powyżej 3 000 000 PLN)
df_filtered = df[df["Price"] <= 3_000_000]

# Wybór mieszkań trzypokojowych
three_room_apartments = df_filtered[df_filtered["Room number"] == 3]

# Obliczenie średniej ceny mieszkań trzypokojowych w każdej dzielnicy
average_price_three_room_per_district = three_room_apartments.groupby("District")["Price"].mean().reset_index()

# Sortowanie wyników dla lepszej czytelności
average_price_three_room_per_district = average_price_three_room_per_district.sort_values(by="Price", ascending=False)

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
bars = plt.barh(average_price_three_room_per_district["District"], average_price_three_room_per_district["Price"], color='skyblue', edgecolor='black')

# Dodanie etykiet na słupkach
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f"{int(bar.get_width()):,}".replace(",", " "), ha='left', va='center', fontsize=8)

# Etykiety i tytuł wykresu
plt.xlabel("Średnia cena mieszkań trzypokojowych (PLN)")
plt.ylabel("Dzielnica")
plt.title("Średnia cena mieszkań trzypokojowych w każdej dzielnicy Poznania")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='x')  # Wyłączenie notacji naukowej dla cen

# Wyświetlenie wykresu
plt.show()
```
